```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
pagelink="https://autoportal.com/newcars/"
```

```python
page_resp=requests.get(pagelink)
```

```python
content=BeautifulSoup(page_resp.content,"html.parser")
```

```python
#based on car names
car_names=[]
hyperlinks=[]
divs=content.find_all("div",{'class':'border_box m_b-20'})
for div in divs:
    link=div.find_all("a")
    lis=div.find_all("li")
    for a in link:
        hyperlinks.append(pagelink+a['href'])
    for b in lis:
        car_names.append(b.text.strip())
```

```python
#based on car type
types=[]
hyperlinks_type=[]
divs=content.find_all("div",{'class':'cell8 cell-md m_b-20'})
for div in divs:
    link=div.find_all("a")
    lis=div.find_all("li")
    for a in link:
        hyperlinks_type.append(pagelink+a['href'])
    for b in lis:
        types.append(b.text.strip())
```

```python
hyperlinks_type_clean = []
types_cleaned=[]
for i in range(9):
    hyperlinks_type_clean.append(hyperlinks_type[i])
    types_cleaned.append(types[i])
```

```python
def get_car_data(link):
    brand_wise_names=[]
    con=requests.get(link)
    content=BeautifulSoup(con.content,'html.parser')
    divs=content.find_all("div",{'class':'model-list__body'})
    for div in divs:
        link=div.find_all("a",{'class':'model-list__title'})
        for a in link:
            brand_wise_names.append(a.text.strip())
    return(brand_wise_names)
    
    
```

```python
cars_dic={}
for i in (range(len(car_names))):
    cars_dic[car_names[i]]=get_car_data(hyperlinks[i])
```
